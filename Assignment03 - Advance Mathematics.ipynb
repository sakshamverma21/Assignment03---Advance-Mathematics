```python
import pandas as pd
import numpy as np
import math
import os
```

```python
df = pd.read_csv("data.csv", encoding='latin1')
```

```python
x = df["no2"].dropna().values
r = 102303892

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)

z = x + ar * np.sin(br * x)
```

```python
mu = np.mean(z)

sigma_sq = np.mean((z - mu) ** 2)

lam = 1 / (2 * sigma_sq)

c = 1 / math.sqrt(2 * math.pi * sigma_sq)
```

```python
print("Estimated Parameters:")
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
